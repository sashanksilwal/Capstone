```python

# # Filter the rows in merged_df
# filtered_df = merged_df[(merged_df['pred'].isin(['ads', 'analytics', 'socials', 'marketing'])) & (merged_df['prob'] > 0.8)]

# Group by url and count occurrences
url_counts = df.groupby('url').size().reset_index(name='count')
filtered_url_counts = classified_thresh_df.groupby('url').size().reset_index(name='count')

# # Sort by count in ascending order
url_counts = url_counts.sort_values(by='count', ascending=True)
filtered_url_counts = filtered_url_counts.sort_values(by='count', ascending=True)

# Add a new column with the cumulative distribution function (CDF) values
url_counts['cdf'] = np.cumsum(url_counts['count']) / url_counts['count'].sum()
filtered_url_counts['cdf'] = np.cumsum(filtered_url_counts['count']) / filtered_url_counts['count'].sum()

# Plot both graphs on the same figure
fig, ax = plt.subplots(figsize=(8, 5))

# Add a grid and set x- and y-axis labels
plt.grid(True, linestyle=':', color='gray', alpha=0.7)
ax.plot(url_counts['count'], url_counts['cdf'], label='Original', linestyle='-', linewidth=2, color='blue')
ax.plot(filtered_url_counts['count'], filtered_url_counts['cdf'], label='Falcon Browser', linestyle='--', linewidth=2, color='red')

ax.set_xlabel('No of JS Requests', fontsize=14)
ax.set_ylabel('CDF', fontsize=14)
ax.set_title('CDF of URL Counts for Predictions with Prob > 0.8', fontsize=16)
ax.legend(fontsize=12)



plt.show()
```

```python
# merge the dataframes on the "url" column
merged_size_df = total_bytes.merge(classified_no_thresh_total_bytes, on="url", how="outer", suffixes=("", "_no_thresh"))
merged_size_df = merged_size_df.merge(classified_thresh_total_bytes, on="url", how="outer", suffixes=("", "_thresh"))

# remove rows with NA values
merged_size_df = merged_size_df.dropna()

# calculate the size differences
merged_size_df["size(bytes)_no_thresh"] = merged_size_df["size(bytes)"] - merged_size_df["size(bytes)_no_thresh"]
merged_size_df["size(KB)_no_thresh"] = merged_size_df["size(KB)"] - merged_size_df["size(KB)_no_thresh"]
merged_size_df["size(MB)_no_thresh"] = merged_size_df["size(MB)"] - merged_size_df["size(MB)_no_thresh"]
merged_size_df["size(bytes)_thresh"] = merged_size_df["size(bytes)"] - merged_size_df["size(bytes)_thresh"]
merged_size_df["size(KB)_thresh"] = merged_size_df["size(KB)"] - merged_size_df["size(KB)_thresh"]
merged_size_df["size(MB)_thresh"] = merged_size_df["size(MB)"] - merged_size_df["size(MB)_thresh"]

# remove unnecessary columns
# merged_df = merged_df.drop(["size(bytes)_no_thresh", "size(KB)_no_thresh", "size(MB)_no_thresh"], axis=1)

merged_size_df
```

```python
import matplotlib.pyplot as plt

# Create a figure with three subplots
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(18,6))

# Define color for the box plot
color = '#8CB4F3'

# Plot the first box plot on the first subplot
bp1 = axs[0].boxplot(merged_size_df["size(KB)"], widths=0.6, patch_artist=True, showfliers=False)
for patch in bp1['boxes']:
    patch.set_facecolor(color)
for median in bp1['medians']:
    median.set(color='black', linewidth=1.5)
axs[0].set_xticklabels([''], fontsize=12)
axs[0].set_ylabel('Size (KB)', fontsize=12)
axs[0].set_title('Original', fontsize=14)
axs[0].tick_params(axis='both', which='major', labelsize=12)
axs[0].grid(axis='y', linestyle='--', alpha=0.7)
axs[0].set_ylim(bottom=0, top=2500)

# Plot the third box plot on the second subplot
bp3 = axs[1].boxplot(merged_size_df["size(KB)_thresh"], widths=0.6, patch_artist=True, showfliers=False)
for patch in bp3['boxes']:
    patch.set_facecolor(color)
for median in bp3['medians']:
    median.set(color='black', linewidth=1.5)
axs[1].set_xticklabels([''], fontsize=12)
axs[1].set_ylabel('Size (KB)', fontsize=12)
axs[1].set_title('Falcon Browser threshold =0.8', fontsize=14)
axs[1].tick_params(axis='both', which='major', labelsize=12)
axs[1].grid(axis='y', linestyle='--', alpha=0.7)
axs[1].set_ylim(bottom=0, top=2500)

# Plot the second box plot on the third subplot
bp2 = axs[2].boxplot(merged_size_df["size(KB)_no_thresh"], widths=0.6, patch_artist=True, showfliers=False)
for patch in bp2['boxes']:
    patch.set_facecolor(color)
for median in bp2['medians']:
    median.set(color='black', linewidth=1.5)
axs[2].set_xticklabels([''], fontsize=12)
axs[2].set_ylabel('Size (KB)', fontsize=12)
axs[2].set_title('Falcon Browser threshold = 0', fontsize=14)
axs[2].tick_params(axis='both', which='major', labelsize=12)
axs[2].grid(axis='y', linestyle='--', alpha=0.7)
axs[2].set_ylim(bottom=0, top=2500)

fig.suptitle('Comparison of Web Page Size Distribution (only due to JS) Requested with and without Falcon Browser, with and without Threshold', fontsize=16)


# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3)

# Show the plot
plt.show()
```
